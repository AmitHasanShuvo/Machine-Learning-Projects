```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pylab as plt 
%matplotlib inline             
from statsmodels.tsa.stattools import adfuller,acf,pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
```

```python
data = pd.read_csv('AirPassengers.csv')
```

```python
data.head()
```

```python
data['Month'] = pd.to_datetime(data['Month'],infer_datetime_format=True)

indexedData = data.set_index(['Month'])
indexedData.head()
```

## Observation
TS deals with two columns. Here, one is month and another is passengers. For easier plotting, we are setting the index o pandas dataframe to the month.
During plots, the index will act as x axis, and other will be at y axis.

```python
# Checking Stationarity

plt.xlabel('Date')
plt.ylabel('Passengers')
plt.plot(indexedData)
```

## Observation

This plotting says that the data is upwards trending. It has different means

```python
# Determining the rolling stats

rolmean = indexedData.rolling(window=12).mean()
rolstd = indexedData.rolling(window=12).std()
#print(rolmean,rolstd)

rolmean.head(20)
```

```python
rolstd.head(20)
```

## Observation

Window size means num of months, giving rolling mean at yearly basis.

```python
# Plotting them

original = plt.plot(indexedData,color = 'blue', label = 'Original')
mean = plt.plot(rolmean,color='red',label = 'Rolling Mean')
std = plt.plot(rolstd, color = 'black',label = 'Rolling std')


plt.legend(loc='best')
plt.title('Rolling mean and std')
plt.show(block = False)
```

## Observation
1. Rolling mean is increasing with time
2. Std is const 

To make this time series stationary, we have to make sure that both mean and std should be time invariant or const with time. 

### Expected 
The mean and std curve should be parallel to x axis. But it's not in our case.

```python
# Performing ADCF test

dftest = adfuller(indexedData['#Passengers'], autolag='AIC')

dfoutput = pd.Series(dftest[0:4],index=['Test Statistics','p-value','#Lags used','Number of Observations used'])

for key,value in dftest[4].items():
    dfoutput ['Criticial value (%s)'%key] = value
print(dfoutput)
```

## Observation

For the TS to be stationary, it's ADCF test should have:
1. p value less than 0.05 (low)
2. The Critical value at 1%, 5%, 10% should be as close as Test statistics

But in this case, it's not satisfying any of these conditions.

## Data Transformation to achieve stationary

Available methods: takings logs, square, sqaure roots, cube, cube roots, exponential decay, time shift etc.

### Log scale transformation

```python
indexedData_logscale = np.log(indexedData)
```

```python
plt.plot(indexedData_logscale)
```

```python
movingAvg = indexedData_logscale.rolling(window=12).mean()
movingstd = indexedData_logscale.rolling(window=12).std()

plt.plot(indexedData_logscale)
plt.plot(movingAvg,color = 'red')
```

## Observation

It's better than previous one, but still not stationary. 

We know that time series with log scale and moving avg have a trend component. We will use elimentary intuition. 

See details at: https://www.kaggle.com/freespirit08/time-series-for-beginners-with-arima

```python
dataLogScaleMinusMovingAvg = indexedData_logscale - movingAvg

dataLogScaleMinusMovingAvg.head(20)
```

```python
dataLogScaleMinusMovingAvg.dropna(inplace = True)
```

```python
dataLogScaleMinusMovingAvg.head()
```

## Test stationary

```python
def test_staionary(timeseries):
    # Rolling statistics
    movingAvg = timeseries.rolling(window =12).mean()
    movingstd = timeseries.rolling(window =12).std()
    
    #Plotting 
    
    original = plt.plot(timeseries, color = 'blue', label = 'Original')
    mean = plt.plot(movingAvg, color = 'red', label = 'Rolling Means')
    std = plt.plot(movingstd, color = 'black', label = 'Rolling std')
    plt.legend(loc='best')
    plt.title('Rolling mean and std')
    plt.show(block = False)
    
    
    # Performing ADCF test

    # Performing ADCF test

    dftest = adfuller(timeseries['#Passengers'], autolag='AIC')

    dfoutput = pd.Series(dftest[0:4],index=['Test Statistics','p-value','#Lags used','Number of Observations used'])

    for key,value in dftest[4].items():
        dfoutput ['Criticial value (%s)'%key] = value
    print(dfoutput)
```

```python
test_staionary(dataLogScaleMinusMovingAvg)
```

## Observation

Now, it's completly fine. 

1. The P value is less than 0.05 (low)
2. The Critical values are clsoe to test statistics 

```python
# To get better accuracy, we will implement exp decay transformation. 


expDecayTran = indexedData_logscale.ewm(halflife=12,min_periods=0, adjust= True).mean()

plt.plot(indexedData_logscale)
plt.plot(expDecayTran, color = 'red')
```

## Observation

But it looks like exp decay is not giving us better advantage. But in stats, inferences cannot be drawn simply by looking at the curves, so let's do ADCF again.

```python
dataLogScaleMinusEXPMovingAvg = indexedData_logscale - expDecayTran

test_staionary(dataLogScaleMinusEXPMovingAvg)
```

## Observation

Now, TS is stationary and p value is good and test stats score is close to critical values

Again trying for better solution by using time shift transformation

```python
dataLogShift = indexedData_logscale - indexedData_logscale.shift()
plt.plot(dataLogShift)
```

```python
dataLogShift.dropna(inplace=True)
test_staionary(dataLogShift)
```

## Observation

Now this is giving us best result according to the visuals but ADCF scores are not as good as previous plot.

We have thus tried out 3 different transformation: log, exp decay & time shift. For simplicity, we will go with the log scale. The reason for doing this is that we can revert back to the original scale during forecasting.

```python
decomposition = seasonal_decompose(indexedData_logscale)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(441)
plt.plot(indexedData_logscale, label = 'original')
plt.legend(loc = 'best')

plt.subplot(412)
plt.plot(trend, label = 'trend')
plt.legend(loc = 'best')


plt.subplot(411)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc='best')

plt.subplot(411)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')

plt.tight_layout()

decomposedLogData = residual
decomposedLogData.dropna(inplace = True)
test_staionary(decomposedLogData)
```
